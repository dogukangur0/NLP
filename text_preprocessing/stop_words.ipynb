```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
```

```python
stop_words_eng = set(stopwords.words("english"))
# set ile hem listele hem de benzersiz olarak ayarla.
stop_words_eng
```

```python
text = "There are some examples of handling stop words from some texts."
```

```python
text_list = text.split()
filtered_words = [word for word in text_list if word.lower() not in stop_words_eng]
filtered_words
```

```python
stop_words_turkish = set(stopwords.words("turkish"))
stop_words_turkish
```

```python
text = "merhaba arkadaşlar çok güzel bir ders işliyoruz."
```

```python
text_list = text.split()
filtered_words = [word for word in text_list if word.lower() not in stop_words_turkish]
filtered_words
```
